```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.decomposition import TruncatedSVD
```

```python
df = pd.read_csv("user-item-ratings.csv")
```

```python
df_pivot = df.pivot_table(values='rating', index='gPlusUserId', columns='gPlusPlaceId', fill_value=0)
df_pivot.head()
```

```python
df_pivot.shape
```

```python
pivot_transpose = df_pivot.T
pivot_transpose.head()
```

```python
normalized_piv = df_pivot.subtract(df_pivot.mean(axis=1), axis = 'rows')
normalized_piv.head()
```

```python
user_user_corr = normalized_piv.T.corr()
user_user_corr.head()
```

```python
picked_userid = '100002870366309555032'
user_user_corr.drop(index=picked_userid, inplace=True)
user_user_corr.head()
```

```python
user_user = user_user_corr[[picked_userid]]
```

```python
user_user.sort_values(picked_userid,ascending=False,inplace=True)
user_user[:15]
```

```python
similar_users = user_user_corr[user_user_corr[picked_userid]>0.4][picked_userid].sort_values(ascending=False)[:15]
```

```python
print(f'top 15 similar users for  {picked_userid} are', similar_users)
```

```python
similar_user_places = normalized_piv[normalized_piv.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_places
```

```python
place_score = {}

for i in similar_user_places.columns:

    place_rating = similar_user_places[i]
    total = 0
    count = 0
    for u in similar_users.index:
        if pd.isna(place_rating[u]) == False:
            score = similar_users[u] * place_rating[u]
            total += score
            count +=1
    place_score[i] = total / count
```

```python
place_score = pd.DataFrame(place_score.items(), columns=['place', 'place_score'])
```

```python
score_sorted = place_score.sort_values(by='place_score', ascending=False)
```

```python
m = 5
score_sorted = score_sorted.head(m)
score_sorted.rename(columns={'place':'gPlusPlaceId'},inplace=True)
```

```python
score_sorted = score_sorted.merge(df_places,on='gPlusPlaceId')
```
